```python
import matplotlib.pyplot as plt

# Group the data by month and gender and sum the revenue for each group
monthly_revenue_gender = df.groupby(['Month', 'Customer_Gender'])['Revenue'].sum().unstack()

# Create a line chart to show the trend of revenue over time for each gender
plt.plot(monthly_revenue_gender.index, monthly_revenue_gender['F'], label='Female')
plt.plot(monthly_revenue_gender.index, monthly_revenue_gender['M'], label='Male')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue by Gender')
plt.legend()
plt.show()

df.info()
```
